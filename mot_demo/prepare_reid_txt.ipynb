```python
import os
data_path = '/home/thaipham/horus/data/reid_data1/imgs'
list_imgs_file = os.listdir(data_path)
list_imgs_file.sort(key=lambda x: int(x[4:]))

train_txt = '/home/thaipham/horus/data/reid_data1/meta/train.txt'
f = open(train_txt, 'w')

for idx, folder_name in enumerate(list_imgs_file):
    folder_path = os.path.join(data_path, folder_name)
    if not os.path.isdir(folder_path):
        continue

    image_files = [f for f in os.listdir(folder_path) if f.lower().endswith(('.jpg'))]

    image_paths = [f'{folder_name}/{image_file} {str(idx)}\n' for image_file in image_files]
    f.writelines(image_paths)
    print(image_paths)
```
